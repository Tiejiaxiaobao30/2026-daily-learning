练习4：定义两个浮点数变量，price 赋值为 19.9, discount 赋值为 0.8 (表示 8 折)。计算折扣后的价格，并将结果存储在变量 final_price 中；计算节省了多少钱，存储在变量 saved_amount 中。然后，使用 f-string 分两行打印出类似以下格式的信息:

练习1：定义两个字符串变量，str1 赋值为 "Hello",str2 赋值为 "Python"。
1. 将这两个字符串拼接起来 (中间加一个空格), 存储在变量 greeting 中
2. 计算 greeting 字符串的长度，存储在变量 length 中
3. 获取 greeting 的第一个字符、第二个字符和最后一个字符

```python
str1='hello'
str2='python' 
greeting=f'{str1} {str2}'
length=len(greeting)
print(f"拼接结果:{greeting}")
print(f"字符串长度:{length}")
print(f"第一个字符:{greeting[0]}")
print(f"第二个字符:{greeting[1]}")
print(f"最后一个字符:{greeting[-1]}")

```

用格式化字符串时 别漏了引号 f" "  索引是用中括号[] 不是小括号！

练习2：定义一个字符串 text = "PythonProgramming"
1. 提取前 6 个字符 (Python), 存储在变量 part1 中
2. 提取后 11 个字符 (Programming), 存储在变量 part2 中
3. 提取中间的 5 个字符 (nProg), 存储在变量 middle 中
4. 计算 part1 和 part2 的长度之和

```python
text='PythonProgramming'
part1=text[:6]
part2=text[6:]
middle=text [5:10]
print(f"前6个字符:{part1}")
print(f"后11个字符:{part2}")
print(f"中间5个字符:{middle}")
print(f"part1和part2的总长度:{len(part1 + part2)}")
```

取左不取右 计算长度用len（） 不仅仅简单拼接 

练习3：定义以下变量: age_str = "25" height_str = "1.75" score = 95
完成以下任务:
1. 将 age_str 转换为整数，存储在 age 中

练习4：定义以下变量:password = "python123"  min_length = 8
完成以下任务:
1. 检查密码长度是否大于等于 min_length, 存储结果
2. 获取密码的第一个字符和最后一个字符
3. 提取密码的前 6 个字符
4. 检查前 6 个字符是否等于 "python"

```python
password='python123'
min_length=8
print(f"密码长度:{len(password)}")
print(f"密码第一个字符：{password[0]}")
print(f"密码最后一个字符：{password[-1]}")
print(f"密码前6个字符:{password[:6]}")
```

第一问和第四问是"比较"运算的应用

```python
is_length_ok = len(password) >= min_length
print(f"长度符合要求(>=8):{is_length_ok}")
is_python = password[:6] == "python"
print(f"前六个字符是否等于'python':{is_python}")
```

is_length_ok和is_python 都是随意定义的变量 所以后面用的是等号=  再后面一个符号就是比较符号 比较符号左右应该为值 这样输出才为T or F
